```python
### eg.

mpop= [1314,5688,8861,5318,5735,1254]
fpop= [2165,6844,8644,7453,3518,1215]
totalpop= [5151,11258,17851,15058,10254,3000]
age= list(range(20,26) )

plt.plot(age, mpop, label="Male Population")
plt.plot(age, fpop, label="Female Population")
plt.plot(age, totalpop, label="Total Population")
plt.legend()
plt.show()
```

```python
## Stack bar

plt.bar(age, totalpop,label="Total Population")
plt.bar(age, mpop,label="Male Population")
plt.bar(age, fpop, label="Female Population")

plt.legend()
plt.show()
```

```python
#plot + bar

plt.plot(age, totalpop,label="Total Population")
plt.bar(age, mpop,label="Male Population")
plt.bar(age, fpop, label="Female Population")

plt.legend()
plt.show()
```

```python
#pie

piePart = [100,40,50,60,70]
labels = ['Hundred','Fourty','Fifty','Sixty','Seventy']
colorList = ['Green','Blue','Red','Yellow','Purple']
plt.pie(piePart, labels=labels, colors=colorList, wedgeprops={'edgecolor':'black'})

plt.show()
```

```python
## explode

piePart = [100,40,50,60,70]
labels = ['Hundred','Fourty','Fifty','Sixty','Seventy']
colorList = ['Green','Blue','Red','Yellow','Purple']
explodelist= [0,0,0,0.5,0]
plt.pie(piePart, labels=labels, colors=colorList,explode= explodelist, wedgeprops={'edgecolor':'black'})

plt.show()
```
