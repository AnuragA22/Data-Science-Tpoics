```python
m1= pd.read_csv("MegaMart_sales.csv" , header=0)
m1.head(3)
```

```python
m2= pd.read_csv("MegaMart_newsales.csv" , header=0)
m2.head(3)
```

```python
## Merge df and sales value

m3=m1.merge(m2, how="outer")

m3.head(3)
```

```python
m4= m3.groupby(by='Category').sum()
m4
```

```python
## drop duplicates

m5 = m3.drop_duplicates(keep='first')
m5
```

```python
m5.groupby(by='Category').sum()
```

```python
### Best category- sub category

m6= m5.groupby(["Category","Sub-Category"]).sum()
m6.sort_values("Profit", ascending=False)
```

```python
## Invalid orders- 7

m5['Order no']=m5['Order ID'].str.split('-')

m5['Order no']=m5['Order no'].apply(lambda x: x[2])

m5['no_of_digits']=m5['Order no'].apply(lambda x: len(x))

m5['no_of_digits'].value_counts()
```

```python
### invalid order =7 , as 7 orders have less 6 digits only valid order contain 7 digits
```

```python
## Occurence of furniture in top 25 sales

m8=m5.sort_values(by='Sales', ascending=False)
m8=m5.iloc[:25,]

m8['Category'].value_counts()
```

```python
### And operation

m5[(m5.Sales > 250) & (m5.Profit > 50)].sort_values(by="Sales", ascending=False)
```

```python
#### Column manuplation

m9 = m3[m3["Profit"]>0]
m9.head(3)
```

```python
m9=m9.loc[m9.Category=='Technology']

m9['Profit_per_qty']=m9['Profit']/m9['Quantity']

m9.sort_values(by='Profit_per_qty',ascending=True)
```
